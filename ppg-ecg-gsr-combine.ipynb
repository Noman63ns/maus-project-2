```python
import pandas as pd
```

```python
df=pd.read_csv('/kaggle/input/ououjh/ppgecggsr.csv')
```

```python
df
```

```python
x=df.iloc[:,1:58]
```

```python
y=df.iloc[:,58]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred)*100)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
lr=LogisticRegression(C=100.00)
lr.fit(x_train,y_train)
y_predlr=lr.predict(x_test)
sv=SVC(C=100)
sv.fit(x_train,y_train)
y_predsc=sv.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predsc))
print(classification_report(y_test,y_predlr))
print(classification_report(y_test,y_predsc))
```

```python
ext=ExtraTreesClassifier()
gbc=GradientBoostingClassifier()
adc=AdaBoostClassifier(n_estimators=25)
ext.fit(x_train,y_train)
adc.fit(x_train,y_train)
gbc.fit(x_train,y_train)
y_predext=ext.predict(x_test)
y_predgbc=gbc.predict(x_test)
y_predadb=adc.predict(x_test)
```

```python
print(classification_report(y_test,y_predext))
print(classification_report(y_test,y_predgbc))
print(classification_report(y_test,y_predadb))
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],              
    'min_samples_split': [2,4,6],                      
    'min_samples_leaf': [1, 2, 3],                  
                              
}
```

```python
rf=RandomForestClassifier()
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                           cv=5, n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
y_predgrdrf=grid_search.predict(x_test)
```

```python
print(classification_report(y_test,y_predgrdrf))
```
